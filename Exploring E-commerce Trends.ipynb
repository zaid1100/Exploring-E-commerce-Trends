# About Dataset
Exploring E-commerce Trends: A Guide to Leveraging Dummy Dataset

Introduction:
In the world of e-commerce, data is a powerful asset that can be leveraged to understand customer behavior,
improve sales strategies, and enhance overall business performance.
This guide explores how to effectively utilize a dummy dataset generated to simulate various aspects of an e-commerce platform.
By analyzing this dataset, businesses can gain valuable insights into product trends, customer preferences, and market dynamics.

Dataset Overview:
The dummy dataset contains information on 1000 products across different categories such as electronics, 
clothing, home & kitchen, books, toys & games, and more. Each product is associated with attributes such as price,
rating, number of reviews, stock quantity, discounts, sales, and date added to inventory. 
This comprehensive dataset provides a rich source of information for analysis and exploration.

Data Analysis:
Using tools like Pandas, NumPy, and visualization libraries like Matplotlib or Seaborn,
businesses can perform in-depth analysis of the dataset. Key insights such as top-selling products, 
popular product categories, pricing trends, and seasonal variations can be extracted through exploratory data analysis (EDA).
Visualization techniques can be employed to create intuitive graphs and charts for better understanding and communication 
of findings.

```python
# convert date in month and year 

df['DateAdded'] = pd.to_datetime(df['DateAdded'], format='%Y-%m-%d')
df['Year'] =df['DateAdded'].dt.year
df['Month'] =df['DateAdded'].dt.month
```

```python
pd.set_option('display.max_rows', None)
total_Price_by_Product =df.groupby('ProductName')['Price'].sum().reset_index()
print(total_Price_by_Product)
```

```python
df.head()
```

```python
# total price and Rating by Product and price
total_price_and_Rating = df.groupby(['ProductName','Category'])[['Price','Rating']].sum().reset_index()
print(total_price_and_Rating)
```

```python
# total_Num_Reviews & StockQuantity by Product and Categroy

total_Num_Reviews_Stock_Quantity = df.groupby(['ProductName','Category'])[['NumReviews','StockQuantity']].sum().reset_index()
print(total_Num_Reviews_Stock_Quantity)
```

```python
# total discount and sales by Product and category wise

total_sales_and_Discount = df.groupby(['ProductName','Category'])[['Sales','Discount']].sum().reset_index()
print(total_sales_and_Discount)
```

```python
df.head()
```

```python
# all total year wise

total_all_numerical_values = df.groupby(['Year'])[['Price','Rating','NumReviews','StockQuantity','Discount','Sales']].sum().reset_index()
print(total_all_numerical_values)
```

```python
# all total  month wise

total_all_numerical_values = df.groupby(['Month'])[['Price','Rating','NumReviews','StockQuantity','Discount','Sales']].sum().reset_index()
print(total_all_numerical_values)
```

```python
# all total  month wise

total_all_numerical_values = df.groupby(['Year','ProductName'])[['Sales']].sum().reset_index()
print(total_all_numerical_values)
```

```python
# Group by city and sum the sales
product_Sales = df.groupby('ProductName')['Sales'].sum().reset_index()
# Sort by amount in descending order and get the top 10
top_10_product = product_Sales.sort_values(by='Sales', ascending=False).head(10)
# Plotting the bar plot
plt.figure(figsize=(14, 5))
ax=sns.barplot(y='Sales', x='ProductName', data=top_10_product)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Top 10 ProductName with Maximum sales')
plt.ylabel('Total sales')
plt.xlabel('ProductName')
plt.show()
```

```python
# total sales by Product wise
# group by 
# Group by city and sum the sales
product_sales = df.groupby('ProductName')['Sales'].sum().reset_index()
# Sort by amount in descending order and get the top 10
bottom_10_product = product_sales.sort_values(by='Sales', ascending=False).tail(10)
# Plotting the bar plot
plt.figure(figsize=(14, 5))
ax=sns.barplot(y='Sales', x='ProductName', data=bottom_10_product)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Top 10 product with Minimum sales')
plt.ylabel('Total sales')
plt.xlabel('productName')
plt.show()
```

```python
# total sales sum by year wise

total_sales =df.groupby('Year')['Sales'].sum().reset_index()
ax=sns.barplot(data=total_sales,x='Year',y='Sales')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total sales sum by productName and Which Year

total_sales =df.groupby(['ProductName','Year'])['Sales'].sum().reset_index()
print(total_sales)
```

```python
# total average price and sales by category wise and year wise
total_price_and_sales = df.groupby(['Category'])[['Price','Sales']].mean().reset_index()
print(total_price_and_sales)
```

```python
# total Rating by Category wise
total_rating = df.groupby(['Category','ProductName'])[['Rating','Price','Sales']].sum().reset_index()
print(total_rating)
```

```python
# total sales by category wise

plt.figure(figsize=(10,7))
total_sales =df.groupby(['Category'])['Sales'].sum().reset_index()
ax=sns.barplot(data=total_sales,y='Category',x='Sales')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total sales by product 
plt.figure(figsize=(10,7))
total_sales =df.groupby(['Category'])['NumReviews'].sum().reset_index()
ax=sns.barplot(data=total_sales,y='Category',x='NumReviews')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('total_Sales')
plt.ylabel('Product_Name')
plt.title('total sales by Product')
```

```python
# Total sales by month and year-wise
plt.figure(figsize=(13,5))
total_sales = df.groupby(['Year','Month'])['Sales'].sum().reset_index()

# Convert Year and Month to strings and concatenate them
total_sales['Year_Month'] = total_sales['Year'].astype(str) + '_' + total_sales['Month'].astype(str)

ax=sns.barplot(data=total_sales,x='Year_Month',y='Sales')
for bars in ax.containers:
    ax.bar_label(bars)
```
